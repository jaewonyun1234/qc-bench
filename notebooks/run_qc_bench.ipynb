## 3) Run the benchmark (Python API)

This does the same thing as the CLI but lets you call it from Python.


```python
from pathlib import Path

from qcbench.runner import run_config

config_path = Path("configs/isw.yaml")
output_path = Path("results/runs.parquet")

df = run_config(config_path, output_path, append=False)
df.head()
```

## 4) Load results

If you ran multiple configs with append=True, the table will contain multiple runs.


```python
import pandas as pd

results = pd.read_parquet("results/runs.parquet")
results.head()
```
